Let's start our adventure by reading the magical Excel file

```python
df = pd.read_excel('Greencell_Danish Ara.xlsx')
```

# Now, it's time to create our plot
 Imagine we're drawing on a giant piece of paper with colorful crayons

 Plot the original data - our first treasure map!


```python
plt.plot(df['Date'], df['Price'], label='Original Price')
```

#Let's create some magic spells called moving averages

```python
# The 10-day moving average is like smoothing the data with a soft cloud
df['MA_10'] = df['Price'].rolling(window=10).mean()
plt.plot(df['Date'], df['MA_10'], label='10-day Moving Average')

# The 50-day moving average is like smoothing it even more, like a big friendly hug
df['MA_50'] = df['Price'].rolling(window=50).mean()
plt.plot(df['Date'], df['MA_50'], label='50-day Moving Average')

# Now, we need to label our masterpiece and summon the legends
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
```
